```python
table = pd.DataFrame(columns=['Metropolitan area', 'rho', 'p_value'])
for i, l in enumerate(lista):
    NT = (df["Tweet search - "+l+"_sentNumber_Negative_sa_det"]).values
    PB = (df["Police_Brutality_lagged_sa_det"]).values
    CR = (df[l+"_crimes_lagged_sa_det"]).values
    corrdf = pd.DataFrame(columns = ["NT(t+1)","NT(t)","PB(t)","CR(t)"], data =  {"NT(t+1)":NT[0:-1],
                                                                                  "NT(t)":NT[1:],
                                                                                  "PB(t)":PB[1:],
                                                                                  "CR(t)":CR[1:]})
    corr = pg.partial_corr(data=corrdf, x='NT(t+1)', y='PB(t)', covar=["NT(t)","CR(t)"], method = "spearman")
    table.loc[len(table.index)] = [listaa[i], round(corr["r"][0],4), round(corr["p-val"][0],4)]
```
